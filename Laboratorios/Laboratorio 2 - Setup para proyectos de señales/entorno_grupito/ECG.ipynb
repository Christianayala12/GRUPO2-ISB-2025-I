```python
# Load NeuroKit and other useful packages
import neurokit2 as nk
import numpy as np
import pandas as pd
import seaborn as sns

# Alternate heart rate and noise levels
ecg180 = nk.ecg_simulate(duration=10, noise=0.05, heart_rate=180)
ecg180_simple = nk.ecg_simulate(duration=10, noise=0.05, heart_rate=180, method="simple")
ecg180_comp = nk.ecg_simulate(duration=10, noise=0.05, heart_rate=180, method="ecgsyn")

ecg30 = nk.ecg_simulate(duration=10, noise=0.01, heart_rate=30)
ecg30_simple = nk.ecg_simulate(duration=10, noise=0.01, heart_rate=30, method="simple")
ecg30_comp = nk.ecg_simulate(duration=10, noise=0.01, heart_rate=30, method="ecgsyn")

# Visualize
ecg180_df = pd.DataFrame({"ECG_180": ecg180, "ECG_180_simple": ecg180_simple, "ECG_180_complejo": ecg180_comp})

ecg30_df = pd.DataFrame({"ECG_30": ecg30, "ECG_30_simple": ecg30_simple, "ECG_30_complejo": ecg30_comp})


nk.signal_plot(ecg180_df, subplots=True)
nk.signal_plot(ecg30_df, subplots=True)
```
