```python
# Load NeuroKit and other useful packages
import neurokit2 as nk
import numpy as np
import pandas as pd
import seaborn as sns

# Simulate
emg8 = nk.emg_simulate(duration=10, burst_number=8, burst_duration=1.0)
emg3 = nk.emg_simulate(duration=10, burst_number=3, burst_duration=1.0)

# Visualize
emg_df = pd.DataFrame({"EMG8": emg8,
                       "EMG3": emg3})
nk.signal_plot(emg_df,subplots=True)
```
